```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def sparsity_measure(vector):
    # Max sparsity = 1 (single 1 in the vector)
    v = np.sort(np.abs(vector))
    n = v.shape[0]
    k = np.arange(n) + 1
    l1norm = np.sum(v)
    summation = np.sum((v / l1norm) * ((n - k + 0.5) / n))
    return 1 - 2 * summation
```

```python
b_dat = np.load("../data/basic.npz")
p_dat = np.load("../data/pos_bcm.npz")
n_dat = np.load("../data/neg_bcm.npz")
```

```python
print(p_dat.keys())
print(p_dat["enc"].shape)
print(p_dat["rec"].shape)
```

```python
enc = b_dat["enc"]

for n_i in range(enc.shape[1]):
    plt.plot(enc[:, n_i])
```

```python
enc[0].shape
```

```python
print(
    "Starting sparsity: {0}".format(sparsity_measure(enc[0][1])))
print("Ending sparsity: {0}".format(sparsity_measure(enc[-1][1])))
```
