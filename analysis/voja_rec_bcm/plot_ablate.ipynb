```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from nengo_learn_assoc_mem.paths import data_path

%matplotlib inline
```

```python
df = pd.read_hdf(os.path.join(data_path, "neg_voja_rec_test", "more_ablate.h5"), "conf", mode="r")
```

```python
plt.figure(figsize=(10, 4))
gg = sns.barplot(x="rec_syn", y="rn_dist", hue="dead_num", data=df[df.noise_mag == 0.0])
gg.set_ylabel("Distance from Runner-up")
gg.set_xlabel("Recurrent Synapse Tau")
plt.legend(loc=(1, 0.2), title="Neurons Ablated")
plt.title("Confidence in Response to Clean Inputs")
```

```python
plt.figure(figsize=(10, 4))
gg = sns.barplot(x="rec_syn", y="rn_dist", hue="dead_num", data=df[df.noise_mag == 0.1])
gg.set_ylabel("Distance from Runner-up")
gg.set_xlabel("Recurrent Synapse Tau")
plt.legend(loc=(1, 0.2), title="Neurons Ablated")
plt.title("Confidence in Response to Noisy Inputs")
```

```python
gg = sns.barplot(x="rec_syn", y="cor", hue="dead_num", data=df[df.noise_mag == 0.])
gg.set_ylabel("Error Rate")
gg.set_xlabel("Recurrent Synapse Tau")
plt.title("Error Rate Across Trials")
```
