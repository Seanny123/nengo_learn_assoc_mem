```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from nengo_learn_assoc_mem.paths import data_path

%matplotlib inline
```

```python
df = pd.read_hdf(os.path.join(data_path, "neg_voja_rec_test", "test_all.h5"), "conf", mode="r")
```

```python
plt.figure(figsize=(10, 4))
gg = sns.barplot(x="rec_syn", y="rn_dist", hue="noise_mag", data=df)
gg.set_ylabel("Distance from Runner-up")
gg.set_xlabel("Recurrent Synapse Tau")
plt.legend(loc=(1, 0.2), title="Noise Magnitude")
plt.title("Confidence in Response to Noisy Inputs")
```

```python
gg = sns.barplot(x="rec_syn", y="cor", hue="noise_mag", data=df)
gg.set_ylabel("Error Rate")
gg.set_xlabel("Recurrent Synapse Tau")
plt.title("Error Rate Across Trials")
```
